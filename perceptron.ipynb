```python
import pandas as pd

df = pd.read_csv('Social_Network_Ads.csv')

df['Male'] = 0
df['Female'] = 0

df.loc[df['Gender'] == 'Male', 'Male'] = 1
df.loc[df['Gender'] == 'Female', 'Female'] = 1

df.drop('Gender', axis=1, inplace=True)
```

```python
df = df[["User ID", "Age", "Male", "Female", "EstimatedSalary", "Purchased"]]
```

```python
df.drop('User ID', axis=1, inplace=True)
```

```python
import numpy as np
import umap
import matplotlib.pyplot as plt

# Generate some sample data
np.random.seed(0)
data = np.random.rand(100, 5)  # 100 data points with 5 features

# Create a UMAP reducer
reducer = umap.UMAP()

# Fit UMAP to the data and transform it to lower dimensions
embedding = reducer.fit_transform(data)

# Plot the reduced data
plt.scatter(embedding[:, 0], embedding[:, 1])
plt.title('UMAP Projection')
plt.show()
```
